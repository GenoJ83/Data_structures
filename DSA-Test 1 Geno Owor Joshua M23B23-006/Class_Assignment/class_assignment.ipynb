```python
#1
import math

def degrees_to_radians(degrees):
    return degrees * (math.pi / 180)

degree = 15
radian = degrees_to_radians(degree)
print(f"Degree: {degree}")
print(f"Radian: {radian}")
```

```python
#2
def parallelogram_area(base, height):
    return base * height

base = 5
height = 6
area = parallelogram_area(base, height)
print(f"Length of base: {base}")
print(f"Height of parallelogram: {height}")
print(f"Area is: {area}")
```

```python
#3
def find_smallest_multiple_with_factors(n):
    factors = []
    result = 1
    for i in range(n, 1, -1):
        factors.append(i)
        result *= i
    return factors, result

n = 13
factors, smallest_multiple = find_smallest_multiple_with_factors(n)
print(f"Factors: {factors}")
print(smallest_multiple)
```

```python
#4
import numpy as np

def test_complex_real_scalar(arr, num):
    complex_check = np.iscomplex(arr)
    real_check = np.isreal(arr)
    scalar_check = np.isscalar(num)
    return complex_check, real_check, scalar_check

arr = np.array([1, 2.5, 3 + 4j])
num = 5

complex_result, real_result, scalar_result = test_complex_real_scalar(arr, num)

print(f"Complex Numbers: {complex_result}")
print(f"Real Numbers: {real_result}")
print(f"Is {num} a Scalar? {scalar_result}")

```

```python
import numpy as np

def get_array_memory_size(arr):
    return arr.nbytes

arr = np.array([1, 7, 13, 105])

size_in_bytes = get_array_memory_size(arr)

print(f"Array: {arr}")
print(f"Memory Size (in bytes): {size_in_bytes}")
```

```python
#2

# List of Catholic Martyrs
catholic_martyrs = [
    "Achileo Kiwanuka",
    "Adolphus Ludigo Mukasa",
    "Ambrosius Kibuuka",
    "Anatoli Kiriggwajjo",
    "Andrew Kaggwa",
    "Antanansio Bazzekuketta",
    "Bruno Sserunkuuma",
    "Charles Lwanga",
    "Denis Ssebuggwawo Wasswa",
    "Gonzaga Gonza",
    "Gyavira Musoke",
    "Yowana Mukiibi",
    "Yusufu Lugalama",
    "Zakayo Lubega",
    "James Buuzaabalyaawo",
    "John Maria Muzeeyi",
    "Joseph Mukasa Kizito",
    "Lukka Baanabakintu",
    "Matiya Mulumba",
    "Mbaga Tuzinde",
    "Mugagga Lubowa",
    "Mukasa Kiriwawanvu",
    "Nowa Mawaggali",
    "Ponsiano Ngondwe"
]

# List of Anglican Martyrs
anglican_martyrs = [
    "Aaron Lubega",
    "Apollo Kivebulaya",
    "Eria Sebukyala",
    "Fredrick Kizza",
    "George Kizza",
    "James Hannington",
    "Janani Luwum",
    "Joseph Balikuddembe",
    "Mark Kakumba",
    "Nuhu Mbegu",
    "Robert Munyagayirwa",
    "Samwiri Mukasa",
    "Yefusa Namayanja",
    "Yohana Mukasa",
    "Yosefu Lugalama",
    "Yowana Kitaka",
    "Yowana Maria Mukasa"
]

# Convert the lists to sets to remove duplicates
catholic_martyrs_set = set(catholic_martyrs)
anglican_martyrs_set = set(anglican_martyrs)

# Find the names that are in both sets (duplicates)
duplicates = catholic_martyrs_set.intersection(anglican_martyrs_set)

# Remove the duplicates from both sets
catholic_martyrs_set -= duplicates
anglican_martyrs_set -= duplicates

# Convert the sets back to lists
catholic_martyrs = list(catholic_martyrs_set)
anglican_martyrs = list(anglican_martyrs_set)

# Combine the lists to create the full list of martyrs
full_list_of_martyrs = catholic_martyrs + anglican_martyrs

print(full_list_of_martyrs)

```

```python
#3
def martyr_count(martyr_name):
    catholic_martyrs = [
        "Achileo Kiwanuka",
        "Adolphus Ludigo Mukasa",
        "Ambrosius Kibuuka",
        "Anatoli Kiriggwajjo",
        "Andrew Kaggwa",
        "Antanansio Bazzekuketta",
        "Bruno Sserunkuuma",
        "Charles Lwanga",
        "Denis Ssebuggwawo Wasswa",
        "Gonzaga Gonza","Kizito"
        "Gyavira Musoke",
        "Yowana Mukiibi",
        "Yusufu Lugalama",
        "Zakayo Lubega",
        "James Buuzaabalyaawo",
        "John Maria Muzeeyi",
        "Joseph Mukasa Kizito",
        "Lukka Baanabakintu",
        "Matiya Mulumba",
        "Mbaga Tuzinde",
        "Mugagga Lubowa",
        "Mukasa Kiriwawanvu",
        "Nowa Mawaggali",
        "Ponsiano Ngondwe"
    ]
    
    anglican_martyrs = [
        "Aaron Lubega",
        "Apollo Kivebulaya",
        "Eria Sebukyala",
        "Fredrick Kizza",
        "George Kizza",
        "James Hannington",
        "Janani Luwum",
        "Joseph Balikuddembe",
        "Mark Kakumba",
        "Nuhu Mbegu",
        "Robert Munyagayirwa",
        "Samwiri Mukasa",
        "Yefusa Namayanja",
        "Yohana Mukasa",
        "Yosefu Lugalama",
        "Yowana Kitaka",
        "Yowana Maria Mukasa"
    ]

    if martyr_name in catholic_martyrs:
        return "Catholic"
    elif martyr_name in anglican_martyrs:
        return "Anglican"
    else:
        return "Unknown"  # In case the name is not found in either group

# Prompt the user to enter a martyr's name
martyr_name = input("Enter a martyr's name: ")

# Use the martyr_count function to determine the group
group = martyr_count(martyr_name)

if group == "Unknown":
    print(f"The martyr '{martyr_name}' is not found in either group.")
else:
    print(f"{martyr_name} belongs to the {group} group.")


```
